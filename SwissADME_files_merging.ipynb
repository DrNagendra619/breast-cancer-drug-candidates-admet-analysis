```python
import pandas as pd

# Load the 3 SwissADME result files
print("Loading SwissADME batch files...")
df1 = pd.read_csv('swissadme_1.csv')
df2 = pd.read_csv('swissadme_2.csv')
df3 = pd.read_csv('swissadme_3.csv')

# Combine them into one master dataset
master_admet = pd.concat([df1, df2, df3], ignore_index=True)

# Save the final consolidated file
master_admet.to_csv('Master_ADMET_Results.csv', index=False)

print(f"✅ Success! Master file created.")
print(f"Total druggable small molecules successfully profiled: {len(master_admet)}")
```
